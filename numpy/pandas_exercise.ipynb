Handling the "Type 2" Nulls
Since we are doing aggregation, NaN values in Type 2 will exclude those Pokemon from multi-group analysis. Fix it immediately

```python
df['Type 2'] = df ['Type 2'].fillna('None')
print(df.to_string())
```

```python

df['Type 2']= df['Type 2'].fillna('NONE')
print(df.head())
```

```python
df.describe()
```

```python
stats_col = ['HP','Attack','Sp. Atk','Sp. Def','Speed','Generation']
df['Total']=df[stats_col].sum(axis = 1)
print(df[['Total']].to_string())
```

```python
type_averages=df.groupby('Type 1')[['Attack','HP','Speed']].mean()
print(type_averages.sort_values('Attack',ascending = False))
```

```python
# Move 'Name' back to a column so .agg({'Name': 'count'}) works
df_reset = df.reset_index()

# Now your original logic will work
type_stats = df_reset.groupby('Type 1').agg({
    'Attack': ['mean', 'max'],
    'Name': 'count'
})

print(type_stats)
```
