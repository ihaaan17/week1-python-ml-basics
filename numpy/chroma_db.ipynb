```python
import chromadb
```

```python
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.get_or_create_collection(name="documents")
```

```python
collection.add(
    documents=[
        "this document is about pineapple",
        "this document is about orange"
        ],
        ids=["id1","id2"]
)
```

```python
from pprint import pprint
results=collection.query(
    query_texts=["this is a document about mumbai"],
    n_results=2,
    where_document={'$contains':'pineapple'}
)
pprint(results)
```
